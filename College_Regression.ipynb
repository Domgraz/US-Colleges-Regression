```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('2022 US College Rankings.csv')
df.head()
```

```python
X = df[['Tuition','Enrollment Numbers']].values
Y = df['Adjusted Rank'].values
```

```python
X_train,X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.175, random_state= 0)
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
y_predict = model.predict(X_test)
print('Predicted Value: ', y_predict)
```

```python
predict_df = pd.DataFrame(X_test, Y_test)
predict_df.head()
```

```python
predict_df = pd.merge(predict_df,df[['College Name','Adjusted Rank']],left_index = True,right_index = True)
predict_df = predict_df.rename(columns = {0: 'Yearly Tuition',1:'Enrollment Numbers','Adjusted Rank': 'Actual Rank','College Name': 'College Name'})
predict_df.head()
```

```python
predict_df['Predicted Rank'] = y_predict
predict_df['Predicted Rank'] = predict_df['Predicted Rank'].round()
predict_df['Difference'] = predict_df['Predicted Rank'] - predict_df['Actual Rank']
predict_df.sort_values(by = 'Difference',ascending=True)
```

```python
mse = mean_squared_error(Y_test,y_predict)
r2 = r2_score(Y_test,y_predict)

print('Mean Squared Error: ',mse)
print('R-Squared: ',r2)
```
